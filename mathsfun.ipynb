```python
import math
import matplotlib.pyplot as plt
```

```python
def sigmoidfun(i):
    return 1/(1+math.exp(-i))

def tanhfun(i):
    return (math.exp(i)-math.exp(-i))/(math.exp(i)+math.exp(-i))

def relufun(i):
    return max(0,i)

def leakyrelufun(i):
    return max(0.1*i,i)

xval=[]
yval=[]
for i in range(-5,5):
    xval.append(i)
    y= leakyrelufun(i)
    yval.append(y)

plt.plot(xval,yval)
```
